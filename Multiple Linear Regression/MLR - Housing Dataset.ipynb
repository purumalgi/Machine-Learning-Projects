# Data Preparation

```python
varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
```

```python
house[varlist] = house[varlist].apply(lambda x: x.map({'yes':1, 'no':0}))
house[varlist].head()
```

```python
# dummy variables

status = pd.get_dummies(house['furnishingstatus'], drop_first=True)
status.head()
```

```python
house = pd.concat([house, status], axis=1)
house.head()
```

```python
house = house.drop('furnishingstatus', axis=1)
house.head()
```

```python
# train test split

df_train, df_test = train_test_split(house, train_size=0.7, random_state=100)
print(df_train.shape)
print(df_test.shape)
```

```python
# rescaling features

scaler = MinMaxScaler()

nums = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

df_train[nums] = scaler.fit_transform(df_train[nums])
df_train.head()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_train.corr(), annot=True, cmap='PuBuGn')
```

# Building a Model

```python
df_train.head()
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
# add a constant
X_train_sm = sm.add_constant(X_train['area'])

# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.params
```

```python
lr_model.summary()
```

```python
# add other variable

X_train_sm = X_train[['area', 'bathrooms']]
X_train_sm = sm.add_constant(X_train_sm)
```

```python
# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# add other variable

X_train_sm = X_train[['area', 'bathrooms', 'bedrooms', 'airconditioning', 'stories']]
X_train_sm = sm.add_constant(X_train_sm)
```

```python
# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

## Adding all variables and building a Model

```python
X_train_sm = sm.add_constant(X_train)
```

```python
# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# Create a dataframe that will contain the names of all feature variables and their respective VIF
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range (X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
X = X_train
X = X_train.drop('semi-furnished', axis=1)
X_train_sm = sm.add_constant(X)
```

```python
# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
X = X.drop('bedrooms', axis=1)

X_train_sm = sm.add_constant(X)

# create linear regression model
lr=sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

# Residual Analysis

```python
y_train_pred = lr_model.predict()
```

```python
res = y_train - y_train_pred
```
